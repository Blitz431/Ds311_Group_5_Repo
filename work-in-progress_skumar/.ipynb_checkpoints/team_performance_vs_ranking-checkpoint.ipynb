```python
# Connecting the fifa_world_cup database
import sqlite3
import pandas as pd

conn = sqlite3.Connection("fifa_world_cup.db")
```

```python
# Selecting necessary columns
fifa_df = pd.read_sql_query('''
    SELECT date, home_team, home_team_fifa_rank, home_team_total_fifa_points, away_team, away_team_fifa_rank, away_team_total_fifa_points 
    FROM fifa_world_cup 
    ORDER BY home_team_fifa_rank ASC, away_team_fifa_rank ASC, home_team_total_fifa_points DESC, away_team_total_fifa_points DESC
        ''',conn)
```
