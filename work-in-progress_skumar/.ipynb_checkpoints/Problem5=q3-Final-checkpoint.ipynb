```python
import sqlite3
import matplotlib.pyplot as plt
from collections import defaultdict

def get_top_winning_streaks(db_path, top_n=10):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Fetch all match results ordered by date
    cursor.execute("""
        SELECT Date, Home_team, Away_team, home_team_score, away_team_score
        FROM fifa_world_cup
        ORDER BY Date ASC
    """)
    
    matches = cursor.fetchall()
    conn.close()

    # Dictionary to track current winning streak per team
    streaks = defaultdict(int)
    max_streaks = defaultdict(int)

    # Iterate through matches to update streaks
    for date, home_team, away_team, home_score, away_score in matches:
        if home_score > away_score:  # Home team wins
            winner, loser = home_team, away_team
        elif away_score > home_score:  # Away team wins
            winner, loser = away_team, home_team
        else:  # Draw
            winner, loser = home_team, away_team  # Both continue their streaks

        # Update winning streaks
        streaks[winner] += 1
        max_streaks[winner] = max(max_streaks[winner], streaks[winner])

        streaks[loser] = 0  # Reset losing team streak

    # Step 1: Get the actual top N teams with the longest winning streaks (sorted descending)
    top_teams = sorted(max_streaks.items(), key=lambda x: x[1], reverse=True)[:top_n]

    # Step 2: Re-sort those top N teams in **increasing** order
    top_teams = sorted(top_teams, key=lambda x: x[1])  

    # Extract data for visualization
    teams = [team for team, streak in top_teams]
    streak_values = [streak for team, streak in top_teams]

    # Plot the data as a bar chart (in increasing order)
    plt.figure(figsize=(12, 8))  # Increased figure size for better readability
    bars = plt.bar(teams, streak_values, color='green', edgecolor='black')

    # Add labels and title
    plt.xlabel("Teams", fontsize=12)
    plt.ylabel("Winning Streak (Games)", fontsize=12)
    plt.title(f"Top {top_n} Longest Winning Streaks (Increasing Order)", fontsize=14)
    plt.xticks(rotation=30)
    plt.grid(True, linestyle="--", alpha=0.6)

    # Add value labels on top of each bar
    for bar in bars:
        yval = bar.get_height()  # Get the height of each bar
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1,  # Position above the bar
                 str(int(yval)), ha='center', va='bottom', fontsize=10)

    # Show the graph
    plt.show()

# Example usage
db_path = "fifa_world_cup.db"
get_top_winning_streaks(db_path, top_n=10)
```

```python
import sqlite3
import matplotlib.pyplot as plt
from collections import defaultdict
import matplotlib.cm as cm
import numpy as np

def get_top_winning_streaks(db_path, top_n=10):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Fetch all match results ordered by date
    cursor.execute("""
        SELECT Date, Home_team, Away_team, home_team_score, away_team_score
        FROM fifa_world_cup
        ORDER BY Date ASC
    """)
    
    matches = cursor.fetchall()
    conn.close()

    # Dictionary to track current winning streak per team
    streaks = defaultdict(int)
    max_streaks = defaultdict(int)

    for date, home_team, away_team, home_score, away_score in matches:
        if home_score > away_score:
            winner, loser = home_team, away_team
        elif away_score > home_score:
            winner, loser = away_team, home_team
        else:
            winner, loser = home_team, away_team  # Both end if draw

        streaks[winner] += 1
        max_streaks[winner] = max(max_streaks[winner], streaks[winner])
        streaks[loser] = 0

    # Get top N teams with longest winning streaks
    top_teams = sorted(max_streaks.items(), key=lambda x: x[1], reverse=True)[:top_n]
    top_teams = sorted(top_teams, key=lambda x: x[1])  # sort in increasing order

    teams = [team for team, streak in top_teams]
    streak_values = [streak for team, streak in top_teams]

    # Generate distinct colors using tab10 colormap
    cmap = cm.get_cmap('tab10', top_n)
    colors = [cmap(i) for i in range(top_n)]

    # Plot the bar chart with individual colors
    plt.figure(figsize=(12, 8))
    bars = plt.bar(teams, streak_values, color=colors, edgecolor='black')

    plt.xlabel("Teams", fontsize=12)
    plt.ylabel("Winning Streak (Games)", fontsize=12)
    plt.title(f"Top {top_n} Longest Winning Streaks (Increasing Order)", fontsize=14)
    plt.xticks(rotation=30)
    plt.grid(True, linestyle="--", alpha=0.6)

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1,
                 str(int(yval)), ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    plt.show()

# Example usage
db_path = "fifa_world_cup.db"
get_top_winning_streaks(db_path, top_n=10)
```
