```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned FIFA dataset
df = pd.read_excel("/Users/saajay/Desktop/GitHub/Ds311_Group_5_Repo/work-in-progress-Saajay-Singh/fifa_world_cup_cleaned.xlsx")


# Q2= Question 3. Overall Home Team Match Result Stats

overall_results = df['home_team_result'].value_counts(normalize=True) * 100
print("Overall Home Team Results (as %):")
print(overall_results)

# Optional: Bar Plot
overall_results.plot(kind='bar', color=['green', 'gray', 'red'], title='Home Team Match Outcomes')
plt.ylabel("Percentage")
plt.xlabel("Result")
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(True)
plt.show()


# additonal question 2. Home Team Advantage by Continent

continent_results = df.groupby('home_team_continent')['home_team_result'].value_counts(normalize=True).unstack().fillna(0) * 100
print("\n Home Team Advantage by Continent (as %):")
print(continent_results)

# Optional: Stacked Bar Plot
continent_results.plot(kind='bar', stacked=True, figsize=(10, 6), title='Home Team Results by Continent')
plt.ylabel("Percentage")
plt.xlabel("Continent")
plt.legend(title="Result", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

 Is there really such a thing as a home team advantage?
 Answer: Yes, there is evidence of home team advantage. Almost half of all home teams win, while losing is significantly less likely. Possible reasons are Familiarity with the field, climate, and altitude. Local fan support (boosts morale), Less fatigue from travel, Possible referee bias (subconscious)
